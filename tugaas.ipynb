```python

def binary_to_other(binary):


    """Konversi bilangan biner ke desimal, oktal, dan heksadesimal."""


    try:


        decimal = int(binary, 2)


        octal = oct(decimal)


        hexadecimal = hex(decimal)


        print("\nHasil Konversi:")


        print(f"Desimal: {decimal}")


        print(f"Oktal: {octal[2:]}")


        print(f"Heksadesimal: {hexadecimal[2:].upper()}")


    except ValueError:


        print("Input tidak valid. Pastikan Anda memasukkan bilangan biner yang benar.")





def decimal_to_other(decimal):


    """Konversi bilangan desimal ke biner, oktal, dan heksadesimal."""


    try:


        decimal = int(decimal)


        binary = bin(decimal)


        octal = oct(decimal)


        hexadecimal = hex(decimal)


        print("\nHasil Konversi:")


        print(f"Biner: {binary[2:]}")


        print(f"Oktal: {octal[2:]}")


        print(f"Heksadesimal: {hexadecimal[2:].upper()}")


    except ValueError:


        print("Input tidak valid. Pastikan Anda memasukkan bilangan desimal yang benar.")





def octal_to_other(octal):


    """Konversi bilangan oktal ke biner, desimal, dan heksadesimal."""


    try:


        decimal = int(octal, 8)


        binary = bin(decimal)


        hexadecimal = hex(decimal)


        print("\nHasil Konversi:")


        print(f"Biner: {binary[2:]}")


        print(f"Desimal: {decimal}")


        print(f"Heksadesimal: {hexadecimal[2:].upper()}")


    except ValueError:


        print("Input tidak valid. Pastikan Anda memasukkan bilangan oktal yang benar.")





def hexadecimal_to_other(hexadecimal):


    """Konversi bilangan heksadesimal ke biner, desimal, dan oktal."""


    try:


        decimal = int(hexadecimal, 16)


        binary = bin(decimal)


        octal = oct(decimal)


        print("\nHasil Konversi:")


        print(f"Biner: {binary[2:]}")


        print(f"Desimal: {decimal}")


        print(f"Oktal: {octal[2:]}")


    except ValueError:


        print("Input tidak valid. Pastikan Anda memasukkan bilangan heksadesimal yang benar.")





def main():


    """Menu utama untuk memilih jenis konversi."""


    while True:


        print("\nPilih jenis konversi:")


        print("1. Biner ke Desimal, Oktal, dan Heksadesimal")


        print("2. Desimal ke Biner, Oktal, dan Heksadesimal")


        print("3. Oktal ke Biner, Desimal, dan Heksadesimal")


        print("4. Heksadesimal ke Biner, Desimal, dan Oktal")


        print("5. Keluar")





        choice = input("Masukkan pilihan Anda (1-5): ")





        if choice == '1':


            binary = input("Masukkan bilangan biner: ")


            binary_to_other(binary)


        elif choice == '2':


            decimal = input("Masukkan bilangan desimal: ")


            decimal_to_other(decimal)


        elif choice == '3':


            octal = input("Masukkan bilangan oktal: ")


            octal_to_other(octal)


        elif choice == '4':


            hexadecimal = input("Masukkan bilangan heksadesimal: ")


            hexadecimal_to_other(hexadecimal)


        elif choice == '5':


            print("Terima kasih telah menggunakan program ini!")


            break


        else:


            print("Pilihan tidak valid. Silakan coba lagi.")





if __name__ == "__main__":


    main()
```
